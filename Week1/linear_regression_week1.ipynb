```python
def compute_gradient(x_train, y_train, w, b):
  m = x_train.shape[0]
  n = x_train.shape[1]
  #dj_dw = np.zeros(100)
  dj_dw = np.zeros(n)
  dj_db = 0
  x_train=x_train.astype(np.float64)
  for i in range(m):
        f_wb = np.dot(w,x_train[i]) + b
        error = f_wb - y_train[i]
        dj_dw += error * x_train[i]
        dj_db += error

  dj_dw = dj_dw / m
  dj_db = dj_db / m
  return dj_dw , dj_db


def gradient_descent(x_train,y_train,w,b,alpha = 0.01):

  # ---------
    # Your code here

    iterations = 500
    for i in range(iterations):
      dj_dw , dj_db = compute_gradient(x_train,y_train,w,b)
      temp_w = w - (alpha * dj_dw)
      temp_b = b - (alpha * dj_db)
      w=temp_w
      b= temp_b

    # Choose learning rate yourself
  # ---------

    return w,b

#compute_gradient(x_train,y_train,w,b)

#gradient_descent(x_train,y_train,w,b)
```
